```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(2,2,figsize = (10,10), sharex = 'col', sharey = 'row')
plt.rcParams.update({'font.size': 18})
n, x = np.loadtxt("./output/coord1000000000_hw_rand.dat", comments = "#", unpack = True)

xtemp = np.logspace(1,9)

ytemp = np.power(xtemp,-0.5)
ax[0][0].set_title("randn()")
ax[0][0].scatter(n,x,color = 'blue', marker='.', label='randn()')
ax[0][0].axhline(y=np.pi, color='black', ls = 'dashed', label = '$\pi$')
ax[0][0].legend()


ax[1][0].scatter(n,abs(x-np.pi),color = 'blue', marker='.')
ax[1][0].set_xlabel('iteration n')
ax[1][0].set_ylabel('$\delta(n)$')
ax[1][0].set_xscale('log')
ax[1][0].set_yscale('log')
a = 1.5
ax[1][0].plot(xtemp, a*ytemp, color='black', ls = 'dashed', label = '$\dfrac{1}{\sqrt{n}}$')
ax[1][0].legend()


n, x = np.loadtxt("./output/coord1000000000_hw_MT.dat", comments = "#", unpack = True)

ax[0][1].set_title("MT\ngenrand_real1()")
ax[0][1].scatter(n,x,color = 'orange', marker='.', label='MT')
ax[0][1].axhline(y=np.pi, color='black', ls = 'dashed', label='$\pi$')
ax[0][1].legend()


ax[1][1].scatter(n,abs(x-np.pi),color = 'orange', marker='.')
ax[1][1].set_xlabel('iteration n')
ax[1][1].set_ylabel('$\delta(n)$')
ax[1][1].set_xscale('log')
ax[1][1].set_yscale('log')
a = 1.5
ax[1][1].plot(xtemp, a*ytemp, color='black', ls = 'dashed', label = '$\dfrac{1}{\sqrt{n}}$')
ax[1][1].legend()



fig.tight_layout()
fig.savefig('./image/MonteCarlo_pi.png')
```
